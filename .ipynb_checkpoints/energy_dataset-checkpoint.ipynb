```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import MinMaxScaler

# Загрузка данных
file_path = 'energy_dataset.csv'  # Укажите путь к вашему датасету
data = pd.read_csv(file_path)

# --- 1. Обработка и анализ данных ---

# Проверка структуры данных
print("Информация о данных:")
print(data.info())

# Проверка статистики данных
print("\nСтатистика данных:")
print(data.describe())

# Проверка пропусков
print("\nПропуски в данных:")
print(data.isnull().sum())

# Заполнение пропусков методом прямой интерполяции
data = data.fillna(method='ffill')

# Преобразование временной метки
if 'time' in data.columns:
    # Удаляем временную зону, установив UTC
    data['time'] = pd.to_datetime(data['time'], utc=True)
    # Удаляем остаточную информацию о временной зоне
    data['time'] = data['time'].dt.tz_localize(None)
    data.set_index('time', inplace=True)
else:
    raise KeyError("В датасете отсутствует колонка 'time'.")

# Извлечение временных признаков
data['hour'] = data.index.hour
data['day_of_week'] = data.index.dayofweek
data['month'] = data.index.month


# Визуализация временной динамики потребления энергии
plt.figure(figsize=(16, 6))
data['total load actual'].plot(label='Actual Load', color='blue')
data['total load forecast'].plot(label='Forecast Load', color='orange')
plt.title('Total Load (Actual vs Forecast)')
plt.legend()
plt.show()

# Корреляционная матрица
plt.figure(figsize=(14, 10))
sns.heatmap(data.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# --- 2. Подготовка данных ---

# Извлечение временных признаков
data['hour'] = data.index.hour
data['day_of_week'] = data.index.dayofweek
data['month'] = data.index.month

# Выбор целевой переменной и признаков
target = 'total load actual'
features = [
    'hour', 'day_of_week', 'month',
    'generation biomass', 'generation fossil brown coal/lignite',
    'generation fossil gas', 'generation solar', 'generation wind onshore',
    'forecast solar day ahead', 'forecast wind onshore day ahead',
    'total load forecast'
]

# Проверка наличия всех признаков
missing_features = [
    feature for feature in features if feature not in data.columns]
if missing_features:
    raise KeyError(f"Отсутствуют следующие признаки: {missing_features}")

X = data[features]
y = data[target]

# Нормализация данных
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# --- 3. Обучение модели ---

# Разделение данных на тренировочную и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# Линейная регрессия
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
y_pred_lr = lin_reg.predict(X_test)

# Gradient Boosting
gb_model = GradientBoostingRegressor(n_estimators=100, random_state=42)
gb_model.fit(X_train, y_train)
y_pred_gb = gb_model.predict(X_test)

# --- 4. Оценка точности ---

# Функция для расчета метрик


def evaluate_model(y_true, y_pred, model_name):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    print(f"\n{model_name}:")
    print(f"MAE: {mae:.2f}")
    print(f"RMSE: {rmse:.2f}")
    print(f"R²: {r2:.2f}")


# Оценка моделей
evaluate_model(y_test, y_pred_lr, "Linear Regression")
evaluate_model(y_test, y_pred_gb, "Gradient Boosting")
```
